```python
 #Prepare the ligand: generate 3D coordinates, add hydrogens, convert to PDBQT

 # generate 3D and add hydrogens (writes KVD.pdb)
!obabel KVD.sdf -O KVD.pdb --gen3d -h
```

```python
# convert pdb -> pdbqt (AutoDock Vina format)
!obabel KVD.pdb -O KVD.pdbqt
```

```python
## These are the grid cordinates already canculated fron pymol

# center_x, center_y, center_z = -8.013, -33.990, -16.275
# size_x, size_y, size_z = 18.85, 18.80, 15.52
```

```python
# Step 3 – Create the Vina configuration file

%%bash
cat > config.txt << EOF
receptor = cleaned_6NPM.pdbqt
ligand = KVD.pdbqt

center_x = -8.013
center_y = -33.990
center_z = -16.275

size_x = 18.85
size_y = 18.80
size_z = 15.52

exhaustiveness = 8
num_modes = 5
out = 6NPM-KVD.pdbqt
EOF

echo "===== config.txt ====="
cat config.txt
```

```python
# Step 4 – Run AutoDock Vina

!vina --config config.txt
```
