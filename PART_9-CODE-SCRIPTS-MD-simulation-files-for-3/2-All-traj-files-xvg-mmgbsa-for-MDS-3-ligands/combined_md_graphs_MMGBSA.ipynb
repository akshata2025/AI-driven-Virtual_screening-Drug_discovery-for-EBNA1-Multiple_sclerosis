```python
import plotly.graph_objects as go

def load_xvg(file, tmax=None):
    time, rmsd = [], []
    with open(file) as f:
        for line in f:
            if line.startswith(('#', '@')):
                continue
            t, r = line.split()[:2]
            t = float(t)   # ns
            r = float(r)   # nm

            if tmax is not None and t > tmax:
                break

            time.append(t)
            rmsd.append(r)
    return time, rmsd


# ---- Common time window (recommended) ----
TMAX = 25  # ns (use common length)

# ---- Load backbone RMSD files ----
t_cav, rmsd_cav = load_xvg("cavosonstat-rmsd_backbone.xvg", TMAX)
t_dyn, rmsd_dyn = load_xvg("dynasore-rmsd_backbone.xvg", TMAX)
t_kvd, rmsd_kvd = load_xvg("kvd-rmsd_backbone.xvg", TMAX)


# ---- Create plot ----
fig = go.Figure()

fig.add_trace(go.Scatter(
    x=t_cav, y=rmsd_cav,
    mode="lines",
    name="EBNA1 [Cavosonstat]"
))

fig.add_trace(go.Scatter(
    x=t_dyn, y=rmsd_dyn,
    mode="lines",
    name="EBNA1 [Dynasore]"
))

fig.add_trace(go.Scatter(
    x=t_kvd, y=rmsd_kvd,
    mode="lines",
    name="EBNA1 [ KVD (Reference)]"
))


# ---- Layout ----
fig.update_layout(
    title="Backbone RMSD of EBNA1 in Different Ligand States",
    xaxis_title="Time (ns)",
    yaxis_title="RMSD (nm)",
    template="simple_white",
    legend_title="Backbone System",
    width=900,
    height=500
)

fig.show()
```

```python
import plotly.graph_objects as go

def load_rmsf_xvg(file):
    res, rmsf = [], []
    with open(file) as f:
        for line in f:
            if line.startswith(('#', '@')):
                continue
            r, v = line.split()[:2]
            res.append(int(float(r)))
            rmsf.append(float(v))
    return res, rmsf


# ---- Load RMSF files ----
res_cav, rmsf_cav = load_rmsf_xvg("cavosonstat-rmsf_ca.xvg")
res_dyn, rmsf_dyn = load_rmsf_xvg("dynasore-rmsf_ca.xvg")
res_kvd, rmsf_kvd = load_rmsf_xvg("kvd-rmsf_ca.xvg")


# ---- Create plot ----
fig = go.Figure()

fig.add_trace(go.Scatter(
    x=res_cav, y=rmsf_cav,
    mode="lines",
    name="Cavosonstat"
))

fig.add_trace(go.Scatter(
    x=res_dyn, y=rmsf_dyn,
    mode="lines",
    name="Dynasore"
))

fig.add_trace(go.Scatter(
    x=res_kvd, y=rmsf_kvd,
    mode="lines",
    name="KVD (Reference)"
))


fig.update_layout(
    title="C RMSF of EBNA1 in Different Ligand-Bound States",
    xaxis_title="Residue Number",
    yaxis_title="RMSF (nm)",
    template="simple_white",
    legend_title="Ligand",
    width=950,
    height=500
)

fig.show()
```

```python
import pandas as pd
import plotly.graph_objects as go
import re

def load_gyr_xvg(file, tmax_ns=25):
    time_ns = []
    rg = []

    with open(file) as f:
        for line in f:
            if line.startswith(('@', '#')):
                continue
            parts = re.split(r'\s+', line.strip())
            if len(parts) >= 2:
                t_ps = float(parts[0])
                t_ns = t_ps / 1000.0
                if t_ns > tmax_ns:
                    break
                time_ns.append(t_ns)
                rg.append(float(parts[1]))

    df = pd.DataFrame({"Time_ns": time_ns, "Rg": rg})
    return df


# ---- Load files (Backbone Rg) ----
df_cav = load_gyr_xvg("cavosonstat-gyr_backbone.xvg")
df_dyn = load_gyr_xvg("dynasore-gyr_backbone.xvg")
df_kvd = load_gyr_xvg("kvd-gyr_backbone.xvg")


# ---- Compute mean  SD ----
stats = {
    "Cavosonstat": (df_cav["Rg"].mean(), df_cav["Rg"].std()),
    "Dynasore": (df_dyn["Rg"].mean(), df_dyn["Rg"].std()),
    "KVD (Reference)": (df_kvd["Rg"].mean(), df_kvd["Rg"].std())
}


# ---- Create plot ----
fig = go.Figure()

fig.add_trace(go.Scatter(
    x=df_cav["Time_ns"],
    y=df_cav["Rg"],
    mode="lines",
    name="Cavosonstat"
))

fig.add_trace(go.Scatter(
    x=df_dyn["Time_ns"],
    y=df_dyn["Rg"],
    mode="lines",
    name="Dynasore"
))

fig.add_trace(go.Scatter(
    x=df_kvd["Time_ns"],
    y=df_kvd["Rg"],
    mode="lines",
    name="KVD (Reference)"
))


# ---- Annotations (Mean  SD) ----
annotations = []
y_pos = 0.98

for ligand, (mean, std) in stats.items():
    annotations.append(dict(
        x=0.01,
        y=y_pos,
        xref="paper",
        yref="paper",
        text=f"{ligand}: {mean:.3f}  {std:.3f} nm",
        showarrow=False,
        align="left",
        font=dict(size=12)
    ))
    y_pos -= 0.06


# ---- Layout ----
fig.update_layout(
    title="Backbone Radius of Gyration (Rg) of EBNA1 in Different Ligand-Bound States",
    xaxis_title="Time (ns)",
    yaxis_title="Radius of Gyration (nm)",
    template="plotly_white",
    width=950,
    height=520,
    legend_title="Ligand",
    annotations=annotations
)

fig.show()


# ---- Print values (for Results section) ----
for ligand, (mean, std) in stats.items():
    print(f"{ligand}: Average Rg = {mean:.3f}  {std:.3f} nm")
```

```python
import pandas as pd
import plotly.graph_objects as go
import re

def load_hbond_xvg(file, tmax_ns=25):
    time_ns = []
    hbonds = []

    with open(file) as f:
        for line in f:
            if line.startswith(('@', '#')):
                continue
            parts = re.split(r'\s+', line.strip())
            if len(parts) >= 2:
                t_ps = float(parts[0])
                t_ns = t_ps / 1000.0
                if t_ns > tmax_ns:
                    break
                time_ns.append(t_ns)
                hbonds.append(int(float(parts[1])))

    df = pd.DataFrame({"Time_ns": time_ns, "Hbonds": hbonds})
    return df


# ---- Load H-bond files ----
df_cav = load_hbond_xvg("cavosonstat-hb_prot_lig.xvg")
df_dyn = load_hbond_xvg("dynasore-hb_prot_lig.xvg")
df_kvd = load_hbond_xvg("kvd-hb_prot_lig.xvg")


# ---- Compute mean  SD ----
stats = {
    "Cavosonstat": (df_cav["Hbonds"].mean(), df_cav["Hbonds"].std()),
    "Dynasore": (df_dyn["Hbonds"].mean(), df_dyn["Hbonds"].std()),
    "KVD (Reference)": (df_kvd["Hbonds"].mean(), df_kvd["Hbonds"].std())
}


# ---- Create plot ----
fig = go.Figure()

fig.add_trace(go.Scatter(
    x=df_cav["Time_ns"],
    y=df_cav["Hbonds"],
    mode="lines",
    name="Cavosonstat"
))

fig.add_trace(go.Scatter(
    x=df_dyn["Time_ns"],
    y=df_dyn["Hbonds"],
    mode="lines",
    name="Dynasore"
))

fig.add_trace(go.Scatter(
    x=df_kvd["Time_ns"],
    y=df_kvd["Hbonds"],
    mode="lines",
    name="KVD (Reference)"
))


# ---- Annotations (Mean  SD) ----
annotations = []
y_pos = 0.98

for ligand, (mean, std) in stats.items():
    annotations.append(dict(
        x=0.01,
        y=y_pos,
        xref="paper",
        yref="paper",
        text=f"{ligand}: {mean:.2f}  {std:.2f}",
        showarrow=False,
        align="left",
        font=dict(size=12)
    ))
    y_pos -= 0.06


# ---- Layout ----
fig.update_layout(
    title="ProteinLigand Hydrogen Bond Analysis of EBNA1 Complexes",
    xaxis_title="Time (ns)",
    yaxis_title="Number of Hydrogen Bonds",
    template="plotly_white",
    width=950,
    height=520,
    legend_title="Ligand",
    annotations=annotations
)

fig.show()


# ---- Print values (for Results) ----
for ligand, (mean, std) in stats.items():
    print(f"{ligand}: Average H-bonds = {mean:.2f}  {std:.2f}")
```

```python
import plotly.graph_objects as go

# ---- Energy Components ----
components = ['VDW', 'Electrostatic (EEL)', 'Polar Solv (EGB)', 'Non-polar Solv (ESURF)']

cavosonstat = [-20.43, -7.16, 16.16, -2.96]
dynasore    = [-27.76, -25.31, 39.49, -4.28]
kvd         = [-20.91, -3.01, 13.93, -3.35]

# ---- Net Binding Energies ----
total_g = {
    "Cavosonstat": sum(cavosonstat),
    "Dynasore": sum(dynasore),
    "KVD (Reference)": sum(kvd)
}

fig = go.Figure()

# ---- Bar plots ----
fig.add_trace(go.Bar(
    x=components,
    y=cavosonstat,
    name='Cavosonstat',
    text=[f"{v:.2f}" for v in cavosonstat],
    textposition='outside'
))

fig.add_trace(go.Bar(
    x=components,
    y=dynasore,
    name='Dynasore',
    text=[f"{v:.2f}" for v in dynasore],
    textposition='outside'
))

fig.add_trace(go.Bar(
    x=components,
    y=kvd,
    name='KVD (Reference)',
    text=[f"{v:.2f}" for v in kvd],
    textposition='outside'
))

# ---- Horizontal Net Gbind Lines ----
# ---- Horizontal Net Gbind Lines (non-overlapping, paper-ready) ----
colors = {
    "Cavosonstat": "blue",
    "Dynasore": "red",
    "KVD (Reference)": "green"
}

# Small vertical offsets to avoid overlap
y_offsets = {
    "Cavosonstat": 0.0,
    "Dynasore": -1.2,
    "KVD (Reference)": 1.2
}

for ligand, g in total_g.items():
    fig.add_shape(
        type="line",
        xref="paper",
        x0=0.02,
        x1=0.53,
        y0=g + y_offsets[ligand],
        y1=g + y_offsets[ligand],
        line=dict(
            color=colors[ligand],
            width=3,
            dash="dot"
        )
    )

    fig.add_annotation(
        xref="paper",
        x=0.92,
        y=g + y_offsets[ligand],
        text=f"{ligand} Net G bind = {g:.2f} kcal/mol",
        showarrow=False,
        font=dict(size=12, color=colors[ligand]),
        align="left"
    )

    # Zero reference line
fig.add_hline(y=0, line_width=1, line_color="black")

fig.update_layout(
    title=dict(
        text="MM-GBSA Binding Energy Decomposition of EBNA1Ligand Complexes",
        x=0.02,                     # center the title
        font=dict(size=18)
    ),
    xaxis_title="Energy Components",
    yaxis_title=" Energy (kcal/mol)",
    template="plotly_white",
    width=1000,
    height=550,
    legend_title="Ligand"
)

fig.show()


# ---- Print Net Binding Energies ----
for lig, g in total_g.items():
    print(f"{lig}: Net G bind = {g:.2f} kcal/mol")
```
