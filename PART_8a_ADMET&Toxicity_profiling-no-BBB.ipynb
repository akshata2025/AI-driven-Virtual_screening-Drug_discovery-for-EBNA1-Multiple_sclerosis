```python
#  So with docking + these ADMET filters, essentially done a complete in silico ADMET and toxicity profiling, and selecting 4 top candidates is a standard outcome.
# for target EBNA1
```

```python
import pandas as pd
# load
swiss = pd.read_csv("/content/swissadme_results.csv")
print(swiss.head())
len(swiss)
##columns
swiss.columns
```

```python
pkcsm = pd.read_csv("/content/pkcsm1Results.csv" , sep='\t')
print(pkcsm.head())
len(pkcsm)
pkcsm.columns
```

```python
admetlab = pd.read_csv("/content/ADMETLB30-RESULTS.csv")
print(admetlab.head())
len(admetlab)
admetlab.columns
```

```python
# üß© Step 1. Verify they have the same number of compounds
print(len(swiss), len(pkcsm), len(admetlab))
```

```python
## üß© Step 2. Add a row index identifier
swiss['ID'] = range(1, len(swiss)+1)
pkcsm['ID'] = range(1, len(pkcsm)+1)
admetlab['ID'] = range(1, len(admetlab)+1)
```

```python
#  Step 3. Merge by ID (not SMILES)

# First merge SwissADME and pkCSM
merged_1 = pd.merge(swiss, pkcsm, on='ID')

# Merge with ADMETlab
merged_all = pd.merge(merged_1, admetlab, on='ID')

print("Merged shape:", merged_all.shape)
merged_all.head()
```

```python
## save to csv
merged_all.to_csv('merged_data.csv', index=False)
```

```python
#  merged dataset

# 305 compounds successfully merged
# 209 combined descriptors (SwissADME + pkCSM + ADMETlab)
# Ready for filtering and prioritization (good vs poor ADMET/toxicity profiles)
```

```python
#  Step-by-Step: Filter Good ADMET & Toxicity Profiles

# apply the filters in three logical stages:
# 1Ô∏è Drug-likeness (Lipinski + Veber)
# 2Ô∏è ADME properties (absorption, distribution, metabolism, excretion)
# 3Ô∏è Toxicity filters
```

```python
# Step 1. Load datasets
merged_all = pd.read_csv("/content/merged_data.csv")
print(merged_all.head())
len(merged_all)
merged_all.columns
```

```python
#  Step 2. Rename SMILES column (optional)

# Since you have SMILES_x and SMILES_y, let‚Äôs keep only one.

if 'SMILES_x' in merged_all.columns:
    merged_all.rename(columns={'SMILES_x': 'SMILES'}, inplace=True)
```

```python
##  Step 3. Apply Filters

#  (A) Drug-likeness (SwissADME)

filter_druglike = (
    (merged_all['Lipinski #violations'] <= 1) &
    (merged_all['Veber #violations'] <= 1) &
    (merged_all['MW_x'] <= 500)
)
```

```python
#  (B) Absorption

# (from SwissADME + pkCSM)

filter_absorption = (
    (merged_all['GI absorption'] == 'High') &
    (merged_all['Intestinal absorption (human)'] > 30)
)
```

```python
#  (C) Distribution SwissADME

filter_distribution = (
    (merged_all['BBB permeant'] == 'No')  # unless CNS target
)
```

```python
#  (D) Metabolism SwissADME

# (avoid strong CYP inhibition)

filter_metabolism = (
    (merged_all['CYP3A4 inhibitor'] == 'No') &
    (merged_all['CYP2D6 inhibitior'] == 'No')
)
```

```python
#  (E) Excretion pkCSM

# (optional ‚Äî moderate clearance, not too high)

# Filter for moderate clearance (example thresholds)
filter_excretion = (
    (merged_all['Total Clearance'] >= 0.1) &  # not too low
    (merged_all['Total Clearance'] <= 1.0)    # not too high
)
```

```python
#  (F) Toxicity

# (from pkCSM + ADMETlab)

filter_toxicity = (
    (merged_all['AMES toxicity'] == 'No') &
    (merged_all['hERG I inhibitor'] == 'No') &
    (merged_all['Hepatotoxicity'] == 'No')
)
```

```python
# Step 4. Combine all filters

final_filter = (
    filter_druglike &
    filter_absorption &
    filter_distribution &
    filter_metabolism &
    filter_excretion &
    filter_toxicity
)

good_candidates = merged_all[final_filter]
print("Good candidates:", len(good_candidates))
```

```python
#  Step 5. Save shortlisted compounds

good_candidates.to_csv("Good_ADMET_Candidates.csv", index=False)
```
