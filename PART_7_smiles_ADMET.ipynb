```python
## code to extract smiles and its compound_id from "top500_for_docking.csv" data for ADMET profiling.....
## Generated smi formats to load files on swissADME and pkCSM and ADMETlab3.0 webservers for ADMET profiling of top 500 compounds.
## Extrated all 3 results in csv file
```

```python
import pandas as pd

# Read both CSV files
df = pd.read_csv('/content/top_500_for_docking.csv')
df.head()
```

```python
data = pd.read_csv('/content/docking_summary.csv')
data.head()
```

```python
# Extract serial number from the 'Ligand' column in the second file
def extract_serial_number(ligand_name):
    try:
        # Split by underscore and take the first part as serial number
        return int(ligand_name.split('_')[0])
    except:
        return None

# Apply the function to extract serial numbers
data['Serial_Number'] = data['Ligand'].apply(extract_serial_number)
data.head()
```

```python
# Create mapping dictionaries from the first file
name_mapping = {}
smiles_mapping = {}

for idx, row in df.iterrows():
    serial_num = idx + 1  # Assuming position in first file corresponds to serial number
    name_mapping[serial_num] = row['Common_Name']
    smiles_mapping[serial_num] = row['SMILES']

# Map the Common_Name and SMILES to the second file using serial numbers
data['Common_Name'] = data['Serial_Number'].map(name_mapping)
data['SMILES'] = data['Serial_Number'].map(smiles_mapping)
data.head()
```

```python
# Select and reorder the columns you want in the final output
final_df = data[['Serial_Number', 'Common_Name', 'SMILES', 'Ligand', 'Best_Affinity_(kcal/mol)']]
final_df.head()
```

```python
# Filter for strong binders (affinity  -7.0)
filtered_df = final_df[final_df['Best_Affinity_(kcal/mol)'] <= -7.0]
filtered_df.head()
```

```python
# Keep only SMILES, Common_Name and binding hits
result = filtered_df[['SMILES','Common_Name', 'Best_Affinity_(kcal/mol)']]
result.head()
```

```python
# save as csv
result.to_csv('filtered_strong_binders.csv', index=False)
```

```python

import pubchempy as pcp
import pandas as pd

# Load your filtered dataframe
df = pd.read_csv('/content/filtered_strong_binders.csv')

# Function to get CID for a SMILES
def get_cid(smiles):
    try:
        compounds = pcp.get_compounds(smiles, 'smiles')
        if compounds:
            return compounds[0].cid
        else:
            return None
    except:
        return None

# Apply function to your dataframe
df['CID'] = df['SMILES'].apply(get_cid)

# Save new CSV with CIDs
df.to_csv('/content/filtered_strong_binders_with_CID.csv', index=False)

# Check results
print(df.head())
```
