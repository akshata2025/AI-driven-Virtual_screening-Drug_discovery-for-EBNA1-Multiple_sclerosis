```python
from rdkit import Chem
import pandas as pd
```

```python
sdf_file = "/content/open structures.sdf"  # your file path
supplier = Chem.SDMolSupplier(sdf_file)
```

```python
## Extract Common names and SMILES
smiles_list = []
common_name_list = []

for mol in supplier:
    if mol is None:
        continue  # skip invalid molecules
    # SMILES
    smiles = Chem.MolToSmiles(mol)

    # Common name from SDF property
    common_name = mol.GetProp("COMMON_NAME") if mol.HasProp("COMMON_NAME") else "No name"

    smiles_list.append(smiles)
    common_name_list.append(common_name)
```

```python
# Create DataFrame
df_drugbank = pd.DataFrame({"Common_Name": common_name_list, "SMILES": smiles_list})
df_drugbank.head(20)  # view first 20 molecules
```

```python
# Drop rows with empty SMILES
df_drugbank = df_drugbank[df_drugbank['SMILES'] != ""].reset_index(drop=True)
print(df_drugbank.shape)
```
