```python
import pandas as pd

# Load (manually downloaded RAW data from pubchem the CSV)
df = pd.read_csv('pubchem_protacxn_CAD53427_bioactivity_protein.csv')
df.head()
```

```python
# Describe
df.columns
```

```python
# Extract the specific columns
selected_columns = ['Bioactivity_ID', 'Activity', 'Compound_CID', 'Substance_SID']
extracted_data = df[selected_columns].copy()
```

```python
extracted_data
```

```python

# Count active and inactive compounds
activity_counts = df['Activity'].value_counts()
```

```python
activity_counts
```

```python
# Separate active and inactive compounds
active_df = df[df['Activity'] == 'Active']
inactive_df = df[df['Activity'] == 'Inactive']

print(f"Total active compounds available: {len(active_df):,}")
print(f"Total inactive compounds available: {len(inactive_df):,}")
```

```python
# Randomly sample 1000 from each (use random_state for reproducibility)
active_sample = active_df.sample(n=1000, random_state=42)
inactive_sample = inactive_df.sample(n=1000, random_state=42)
```

```python
# Combine the samples
balanced_dataset = pd.concat([active_sample, inactive_sample])
```

```python

# Shuffle the combined dataset
balanced_dataset = balanced_dataset.sample(frac=1, random_state=42).reset_index(drop=True)
```

```python
print(f"\nFinal balanced dataset size: {len(balanced_dataset):,}")
print("Activity distribution in balanced dataset:")
print(balanced_dataset['Activity'].value_counts())
```

```python
# Save the balanced dataset
balanced_dataset.to_csv('balanced_bioactivity_data_1000each.csv', index=False)
print("\nBalanced dataset saved to 'balanced_bioactivity_data_1000each.csv'")

# Display first few rows
print("\nFirst 10 rows of balanced dataset:")
print(balanced_dataset.head(10))
```

```python
# Load your balanced dataset
df = pd.read_csv('balanced_bioactivity_data_1000each.csv')

# Function to get SMILES using connectivity_smiles
def get_smiles(cid):
    try:
        compound = pcp.Compound.from_cid(int(cid))
        return compound.connectivity_smiles  # updated property
    except:
        return None
```

```python
# Apply function to get SMILES
df['SMILES'] = df['Compound_CID'].apply(get_smiles)
```

```python
# Show first few rows
print(df[['Compound_CID', 'SMILES', 'Activity']].head(10))
```

```python
# Save the dataset with SMILES to a new CSV file
df.to_csv('balanced_bioactivity_with_smiles.csv', index=False)

print("Dataset with SMILES saved to 'balanced_bioactivity_with_smiles.csv'")
```
